##  Тождества бухучета

Что должна показать система бухучета - две исходные задачи

1. Состояние предприятия в любой момент времени


- имущество предприятия (активы)
- кто имеет права на это имущество (пассивы)

```
Активы = Пассивы                      (1)
```


2. Результаты работы за пероид

```
Выручка - Затраты = Прибыль           (2)
```

Небольшая сложность подружить эти два уравнения в том, что первое про запасы (остаток величины на момент времени - "сколько воды в ванной"), а второе про поток ("сколько налили и вылили воды из ванной"). Нам приходит на помощь тот факт, что разность между двумя потоками - это запас. 

Расшифруем пассивы:

```
Пассивы = Капитал + Прибыль + Долг    (3)
```

Пока в пассивах не было прибыли, картина выглядела примерно так - учредители внесли имущеcтво в фирму, оформили этот взнос как запись о капитале, привлекли дополнительные заемные средства и оформили это как запись о долге. Почему прибыль вдруг в пассивах?

Прибыль - это увеличение капитала, который инвестировали акционеры. Если фирма увеличила общую величину активов, этот прирост приходится на акционеров. 

Немного поколдовав над переменными мы приходим к следующему виду тождества 1:

```
Денежные средства + Запасы + Счета к получению + Здания, сооружения, оборудование = Капитал + (Доходы текущего периода - Расходы текущего периода) + Счета к оплате + Долг
```
